```python
reads = pd.read_csv('../QIIME2.2/per-sample-fastq-counts.tsv', sep = '\t')
reads = reads[~reads['sample ID'].str.contains('A')]
reads = reads[~reads['sample ID'].str.contains('B')]
print(reads.sum())
print(reads.mean())
print(reads.std())
reads.head()
```

```python
filtering = pd.read_csv('../QIIME2.2/metadata.tsv', sep = '\t', skiprows = [1])
filtering = filtering[~filtering['sample-id'].str.contains('A')]
filtering = filtering[~filtering['sample-id'].str.contains('B')]
print(filtering.sum())
print(filtering.mean())
print(filtering.std())
filtering.head()
```
