```python
import pandas as pd
```

```python
mapping_files = []
num_tabs = 5
for i in range(num_tabs):
    mapping_files.append(pd.read_excel('../QIIME2/mapping_file/bermudagrass_metadata.xlsx', engine='openpyxl', sheet_name = i))
```

```python
mapping_files[2].head()
```

```python
mapping_files_notnull = []
for i in range(len(mapping_files)):
    mapping_files_notnull.append(mapping_files[i][mapping_files[i]['Lab ID'].notnull()])
    mapping_files_notnull[i].set_index('Lab ID', inplace = True)
    mapping_files_notnull[i].sort_index(inplace = True)
    print(mapping_files_notnull[i].shape)
```

```python
full_df = pd.concat([mapping_files_notnull[0], mapping_files_notnull[1], mapping_files_notnull[3], mapping_files_notnull[4]], axis = 1)
full_df.drop(['Unnamed: 5', 'Unnamed: 6','AM-responsiveness = [(dry mass of mycorrhizal plant) – (dry mass of non-mycorrhizal plant) / (dry mass of mycorrhizal plant)] * 100'], inplace = True, axis = 1)
full_df.index.name = '#SampleID'
```
